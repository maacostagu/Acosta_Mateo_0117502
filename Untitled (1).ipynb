```python
#%% DECISIONES Y CICLOS
#1
a = int(input('Ingrese un numero real: '))# Se solicita al usuario que ingrese
#un número real.

if a >= 0: # Si valor ingresado es positivo se imprime el mismo nímero.
    print('El valor absoluto es: ')
    print(a)
if a < 0:
    print('El valor absoluto es: ') # si el valor es negativo se imprime el 
    #valor ingresado multiplicado por -1.
    print(-a)
    
```

```python
#%% 
#2
a = int(input('Ingrese el primer número: '))
b = int(input('Ingrese el segundo número: '))

mag = abs(a - b) # asignamos el valor absoluto de la diferencia entre los 
#números ingresados
print('La magnitud de la diferencia entre los dos números es: ')
print(mag)

if mag%2==0 or mag%3==0 or mag%5==0 or mag%7==0: # condicionamos a que el 
    #residuo de la diferencia de los números primos.
    print(f'{mag} no es primo') # imprime si cumple la condición.
else:
    print(f'{mag} es primo') # Imprime si no se cumple la condición.
    
```

```python
#%% 3                                  
def multiplo(valor, multiplo):
    """
    Funcion para calcular si el numero es multiplo
    utilizando el modulo de la division
    """
    residuo = valor % multiplo   
    if residuo == 0:      #si el resto entre los dos anteriores es 0 significa
        #que uno es multiplo del otro.
        return True
    else:
        return False
 
# lista que contendra los valores multiples.
multiplos_3=[]  
multiplos_4=[]
 
# bucle del 1 al 100.
for i in range(1,101):
 
    if multiplo(i,3):    
        multiplos_3.append(i)  # mediante append agregamos cada numero que 
        #este en el rango solicitado y cumpla la condicion a la lista de 
        #multiplos de 3. 
 
    if multiplo(i,4):
        multiplos_4.append(i)
#se imprimen los resultados.
print ("Los multiplos de 3 son:", multiplos_3)   
print( "")
print( "Los multiplos de 4 son:", multiplos_4)
```
